```python
import torch
from torch.autograd import Variable
import torch.optim as optim
```

```python
# Define the linear model function
def linear_model(x, W, b):
    print(x)
    o = torch.mul(x, W) + b
    return o
```

```python
# Training data and targets
data = torch.arange(1, 20001).reshape(20000, 1).float()
targets = (torch.arange(1, 20001) * 200).reshape(20000, 1).float()



```

```python
# Initialize weights and bias
W = Variable(torch.tensor([190.]), requires_grad=True)
b = Variable(torch.tensor(1.), requires_grad=True)

# Optimizer
optimizer = optim.Adam([W])

# Training loop
for sample, target in zip(data, targets):
    optimizer.zero_grad(set_to_none=True)  # Clear the gradients
    output = linear_model(sample, W, b)  # Forward pass
    loss = (output - target) ** 2  # Compute loss
    loss.backward(torch.ones_like(loss))  # Backward pass (gradient computation)
    optimizer.step()  # Update weights
```

```python

# Test the trained model with a sample input
test_sample = torch.tensor([-100]).float()  # Define a test sample
test_output = linear_model(test_sample, W, b)  # Forward pass with the test sample
print("Test input:", test_sample)
print("Model output for test input:", test_output)
```
